```python
 import seaborn as sns
sns.set_theme(style='whitegrid')


g = sns.catplot(
    data=df, kind="bar",
    x="ParameterName", y="Recall Rate of Matches", hue="Parameter" ,
    ci="sd", palette="Spectral", alpha=.9, height=8, aspect=0.9
)
g.despine(left=True)
g.set_axis_labels("Parameter Combinations ", "Recall Rate for Consecutive/Overlapping Images")
g.legend.set_title("Parameter Combinations")
g.fig.suptitle("recall Rate for ORB for different parameter combinations")
```
