```python
AGAST_SIFT=df1.loc[(df1['Detector/Descriptor'] == "AGAST+SIFT")]
```

```python
AGAST_SIFT
```

```python
AGAST_SIFT1=df2.loc[(df2['Detector/Descriptor'] == "AGAST+SIFT")]
```

```python
AGAST_SIFT1['Parameter'] = "Agast: Threshold= 80; SIFT: nfeatures = 1000, contrastThreshold=0.02, edgeThreshold=4"
```

```python
AGAST_SIFT1
```

```python
ALL_AGAST=pd.concat([AGAST_SIFT, AGAST_SIFT1])
```

```python
ALL_AGAST
```

```python
 import seaborn as sns
sns.set_theme(style='whitegrid')


# Draw a nested barplot by species and sex
g = sns.catplot(
    data=ALL_AGAST, kind="bar",
    x="ParameterName", y="Number of Keypoints", hue="Parameter" ,
    ci="sd", palette="Spectral", alpha=.9, height=8, aspect=0.9
)
g.despine(left=True)
g.set_axis_labels("Parameter Combinations ", "Total Number of Keypoints b/w Consecutive/Overlapping Images")
g.legend.set_title("Parameter Combinations")
g.fig.suptitle("Total Number of Keypoints Detected for AGAST+SIFT for different parameter combinations")
```
